```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Sample dataset
data = pd.read_csv('car_crashes.csv')
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total', y='speeding', data=data)
plt.title('Scatter Plot of Total vs. Speeding')
plt.xlabel('Total')
plt.ylabel('Speeding')
plt.grid(True)
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='abbrev', y='alcohol', data=data.dropna())
plt.title('Bar Plot of Alcohol by Abbrev')
plt.xlabel('Abbrev')
plt.ylabel('Alcohol')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data.drop(columns=['abbrev']))
plt.title('Box Plot of All Numeric Columns')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['ins_premium'].dropna(), kde=True, color='orange')
plt.title('Histogram of Ins Premium')
plt.xlabel('Ins Premium')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(data['speeding'], bins=5, color='orange')
plt.title('Histogram')
plt.xlabel('Speeding')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
